```python
from sklearn.linear_model import LinearRegression
```

```python
def score_with_n_vars(n):
  lin_model = LinearRegression(fit_intercept=True)

  X_train_data_n = X_train_scaled[:, :n]
  X_test_data_n= X_test_scaled[:, :n]
  print(X_train_data_n.shape)
  print(X_test_data_n.shape)
  lin_model.fit(X_train_data_n, y_train)
  
  return [lin_model.score(X_train_data_n, y_train),
          lin_model.score(X_test_data_n, y_test)]
  
```

```python
train_scores, test_scores = [], []
for i in range(1, 287):
  scores = score_with_n_vars(i)
  train_scores.append(scores[0])
  test_scores.append(scores[1])
```

```python
# capturing the nonzero features suggested by the Lasso regression
def non_zero_coeff(alpha):
  lin_model = Lasso(alpha=alpha)
  lin_model.fit(X_train, y_train)
  coefs = lin_model.coef_
  #print(coefs)
  coords = []
  for i in range(len(coefs)):
    if coefs[i] != 0:
      coords.append(i)
  return coords
```
